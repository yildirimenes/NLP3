```python
# https://pypi.org/project/instabot/
import instabot

import pandas as pd
import json
from datetime import datetime
import time
import random
from dotenv import load_dotenv
import os
```

Instagram

```python
load_dotenv()
insta_name = os.getenv("INS_USERNAME")
insta_pwd = os.getenv("INS_PASSWORD")

```

```python
bot = instabot.Bot()
bot.login(username=insta_name, password=insta_pwd)
```

```python
user_id = bot.get_user_id_from_username("leomessi")
```

```python
print(user_id)
```

```python
# tahmini 60 - 150 sn
medias = bot.get_total_user_medias(user_id)
```

```python
type(medias)
```

```python
len(medias)
```

```python
medias[:]
```

```python
leo_posts = []

for i, media_id in enumerate(medias):
    if i % 5 == 0:
        print("Query no:", i)


    try:
        media_info = bot.get_media_info(media_id)[0]


        like_count = media_info["like_count"]

        leo_posts.append({

            "lc": like_count
        })
    except:
        print("Hata!")

    time.sleep(random.randint(2, 10))
```

```python
print(type(leo_posts))
```

```python
df = pd.DataFrame(leo_posts)
df.head()
```

```python
df.to_csv("leo_posts.csv", index=False)
```
